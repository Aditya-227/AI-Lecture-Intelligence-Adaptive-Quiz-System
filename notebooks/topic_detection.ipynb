```python
file_path = "/kaggle/input/datasets/anonymous227/lecture1-transcript-txt/lecture1_transcript.txt"

with open(file_path, "r", encoding="utf-8") as f:
    transcript = f.read()

print(transcript[:1000])
```

```python
import textwrap

chunks = textwrap.wrap(transcript, 500)

print("Total chunks:", len(chunks))
print("\nFirst chunk:\n")
print(chunks[0])
```

```python
from bertopic import BERTopic

topic_model = BERTopic()

topics, probs = topic_model.fit_transform(chunks)
```

```python
topic_model.get_topic_info()
```

```python
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer

vectorizer_model = CountVectorizer(stop_words="english")

topic_model = BERTopic(vectorizer_model=vectorizer_model)

topics, probs = topic_model.fit_transform(chunks)
```

```python
topic_model.get_topic_info()
```

```python
topics = topic_model.get_topics()

for topic_id in topics:
    if topic_id == -1:
        continue
        
    print(f"\nTopic {topic_id}")
    
    words = topic_model.get_topic(topic_id)
    
    for word, score in words[:10]:
        print(word)
```

```python
topics = topic_model.get_topics()

with open("lecture_topics.txt", "w") as f:
    for topic_id in topics:
        if topic_id == -1:
            continue
        
        words = topic_model.get_topic(topic_id)
        keywords = [word for word, score in words[:10]]
        
        f.write(f"Topic {topic_id}: {', '.join(keywords)}\n")
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "google/flan-t5-base"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
import textwrap

summary_chunks = textwrap.wrap(transcript, 1000)

print("Total chunks for summarization:", len(summary_chunks))
```

```python
summaries = []

for chunk in summary_chunks:

    prompt = "Provide a detailed lecture summary of the following content:\n" + chunk

    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=512)

    outputs = model.generate(
        **inputs,
        max_length=200,
        min_length=80,
        num_beams=5,
        no_repeat_ngram_size=2,
        early_stopping=True
    )

    summary = tokenizer.decode(outputs[0], skip_special_tokens=True)

    summaries.append(summary)

print(summaries[0])
```

```python
final_summary = " ".join(summaries)

print(final_summary[:2000])
```

```python
with open("lecture_summary.txt", "w") as f:
    f.write(final_summary)
```

```python
from keybert import KeyBERT

kw_model = KeyBERT()
```

```python
keywords = kw_model.extract_keywords(
    transcript,
    keyphrase_ngram_range=(1, 2),
    stop_words="english",
    top_n=15
)

for word, score in keywords:
    print(word)
```

```python
with open("lecture_concepts.txt", "w") as f:
    for word, score in keywords:
        f.write(word + "\n")
```
